# Importando a base de dados

```python
#Importando Bibliotecas

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Importando base de dados

df = pd.read_csv('https://github.com/MarcosCristofani/Studentperformance/raw/main/StudentPerformanceFactors.csv', sep = ',')
df.head()
```

# Preparando os dados

```python
#Analisando os dados

df.info()
df.isnull().sum()
```

```python
#Renomeando as colunas

df.rename(columns={'Hours_Studied':'Horas_de_estudo',
'Attendance':'Presenca',
'Parental_Involvement':'Engajamento_parental',
'Access_to_Resources':'Acesso_a_recursos',
'Extracurricular_Activities':'Atividades_extracurriculares',
'Sleep_Hours':'Horas_de_sono',
'Previous_Scores':'Pontuacoes_anteriores',
'Motivation_Level':'Nivel_de_motivacao',
'Internet_Access':'Acesso_a_internet',
'Tutoring_Sessions':'Sessoes_de_reforco',
'Family_Income':'Renda_familiar',
'Teacher_Quality':'Qualidade_do_professor',
'School_Type':'Tipo_de_escola',
'Peer_Influence':'Influencia_do_colega',
'Physical_Activity':'Atividade_fisica',
'Learning_Disabilities':'Deficiencia_de_aprendizado',
'Parental_Education_Level':'Nivel_de_formacao_dos_pais',
'Distance_from_Home':'Distancia_da_casa',
'Gender':'Sexo',
'Exam_Score':'Pontuacao_do_exame'},inplace=True)

df.head(2)
```

```python
df.isnull().sum()
```
